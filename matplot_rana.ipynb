# Visualization Choice: Proportions
To show the percentages of replies in the Evaluation variable, broken down by non-white and sex categories, I would like to make a stacked bar chart. We can see how the demographic breakdown influences evaluation replies using this kind of visualisation, which is very helpful when comparing different sets of proportions
## What Makes Bars Stacked?
Comparison of the Proportions: It is simpler to see the relationship between assessment replies and demographics when stacked bars are used to indicate how different categories contribute to the total.
Composition Highlighting: By highlighting the ways in which various groups of people interpret political assessments, this visualisation will help me better grasp the dynamics of public opinion, which is crucial for achieving my professional objectives in political analysis and voter behaviour
## Considering
Clarity: Although stacked bars are helpful, it's crucial to make sure the sections can be easily distinguished. Readability will be improved by using bold colours and unambiguous labelling.
Options: Although I'll probably go with stacked bars, I might also think about side-by-side bars for a direct comparison of proportions since they can help highlight disparities across groups.


```python
data = pd.read_csv('cces_example_data.csv')
data.head()
```

```python
fig, axis = plt.subplots(1, 2, figsize=(12, 6))

axis[0].set_title("Age Histogram")
axis[1].set_title("Political Identity")

plt.show()
```

```python
evaluation_counts = data['Evaluation'].value_counts()

plt.figure(figsize=(8, 5))
evaluation_counts.plot(kind='bar', color='skyblue')
plt.title('Evaluation Responses')
plt.xlabel('Evaluation Categories')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(data['ConMediaUse'], data['ImmiPhobia'], alpha=0.6)
plt.title('Scatter Plot of ConMediaUse vs. ImmiPhobia')
plt.xlabel('ConMediaUse')
plt.ylabel('ImmiPhobia')
plt.grid(True)
plt.show()
```

# Enhancing Your Scatter Plot

There are a few ways to improve the ConMediaUse and ImmiPhobia scatter plot:

Adding Colour Coding: Different colours can be used to stand for distinct categories of a third variable (such as non-white or sex). Finding trends within various demographic groupings might be aided by this (Course Material, Section 3.2).

Including Trend Lines: (Course Material, Section 4.1) Including a regression line can make it easier to see how the two variables are related.

Using Markers: Additional context can be provided by using markers of different sizes or shapes to indicate distinct groups or levels of another variable (Course Material, Section 2.4).

Amplification of Text: The transparency level (alpha) can be changed to reduce overplotting, particularly in cases where the dataset is extensive (Course Material, Section 5.3).
